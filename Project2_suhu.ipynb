```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Membaca file CSV
data = pd.read_csv('suhu permukaan 1.csv')
```

```python
data
```

```python
# Membuat DataFrame dari dataset
df = pd.DataFrame(data)

# Membuat plot menggunakan Seaborn
plt.figure(figsize=(8, 5))
sns.lineplot(data=df, x='Waktu', y='Asimilasi', marker='o', label='Asimilasi', color='blue')
sns.lineplot(data=df, x='Waktu', y='Observasi', marker='o', label='Observasi', color='yellow')
sns.lineplot(data=df, x='Waktu', y='Non Asimilasi', marker='o', label='Non Asimilasi', color='red')

# Mengatur rentang sumbu y
plt.ylim(20, 27)

# Menghilangkan koma pada sumbu y
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: int(x)))

# Menambahkan judul dan label sumbu pada plot
plt.xlabel('Waktu\n(Tanggal)', weight='bold')
plt.ylabel('Celcius', weight='bold')
plt.title('Suhu Permukaan', weight='bold')

# Memutar label sumbu x agar tidak bertumpuk
plt.xticks(rotation=45)

# Menampilkan legenda
plt.legend()

# Menampilkan plot
plt.tight_layout()
plt.show()
```
